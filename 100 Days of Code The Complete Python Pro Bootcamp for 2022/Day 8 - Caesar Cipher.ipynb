```python
#Lista de letras do alfabeto
alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q',
            'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

#Logo do nosso programa
logo = """           
 ,adPPYba, ,adPPYYba,  ,adPPYba, ,adPPYba, ,adPPYYba, 8b,dPPYba,  
a8"     "" ""     `Y8 a8P_____88 I8[    "" ""     `Y8 88P'   "Y8  
8b         ,adPPPPP88 8PP"""""""  `"Y8ba,  ,adPPPPP88 88          
"8a,   ,aa 88,    ,88 "8b,   ,aa aa    ]8I 88,    ,88 88          
 `"Ybbd8"' `"8bbdP"Y8  `"Ybbd8"' `"YbbdP"' `"8bbdP"Y8 88   
            88             88                                 
           ""             88                                 
                          88                                 
 ,adPPYba, 88 8b,dPPYba,  88,dPPYba,   ,adPPYba, 8b,dPPYba,  
a8"     "" 88 88P'    "8a 88P'    "8a a8P_____88 88P'   "Y8  
8b         88 88       d8 88       88 8PP""""""" 88          
"8a,   ,aa 88 88b,   ,a8" 88       88 "8b,   ,aa 88          
 `"Ybbd8"' 88 88`YbbdP"'  88       88  `"Ybbd8"' 88          
              88                                             
              88           
"""

#Definição da função de encriptação.   
def caesar(command, content, change):
    message = ""
    
    #Reduzindo o número a no máximo dois dígitos, de acordo com a quantidade de letras na variável alfabeto.
    change %= 26
    
    #Loop for iniciando verificando se cada letra da mensagem. 
    for letter in content:
        
        #Condicional para verificar se a letra está dentro da variável alphabet. Caso sim, deverá pegar o indice correspondente do alphabet.
        if letter in alphabet:
            ind = alphabet.index(letter)
            
        #Caso não, neste caso é um número ou símbolo, então somente deverá ser adicionado a string message e o programa retornar para o for.
        else:
            message += letter
            continue
        # Condicional para que, de acordo com o encode ou decode adicione ou diminua posições a nossa lista ind com os indexes de acordo com o número dado em chage.
        if command == "encode":
            ind += change
        elif command == "decode":
            ind -= change
        #Condicional para que caso o valor fique maior do que o cumprimento do alfabeto, reduzir ainda mais o número.
        if ind > len(alphabet) - 1:
            ind -= len(alphabet)
        message += alphabet[ind]
    print(f"The decoded text is {message}")

#Início do programa mostrando a logo
print(logo)
restart = True

#Solicitação de informações ao funcionário de direção da criptografia, o texto e quantidade de posições desejada. O programa irá parar a execução quanto o usuário disser que não deseja tentar mais.
while restart:
    direction = input("Type 'encode' to encrypt, type 'decode' to decrypt:\n")
    text = input("Type your message:\n").lower()
    shift = int(input("Type the shift number:\n"))
    caesar(direction, text, shift)
    restart = input("Type 'yes' if you want to go again. Otherwise type 'no'.\n")
    if restart == "no":
        restart = False 
```

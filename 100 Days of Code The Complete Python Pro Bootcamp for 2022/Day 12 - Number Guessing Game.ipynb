```python
#Biblioteca para gerar escolhas aleatórias.
import random
```

```python
logo = """
  / _ \_   _  ___  ___ ___  /__   \ |__   ___    /\ \ \_   _ _ __ ___ | |__   ___ _ __ 
 / /_\/ | | |/ _ \/ __/ __|   / /\/ '_ \ / _ \  /  \/ / | | | '_ ` _ \| '_ \ / _ \ '__|
/ /_\\| |_| |  __/\__ \__ \  / /  | | | |  __/ / /\  /| |_| | | | | | | |_) |  __/ |   
\____/ \__,_|\___||___/___/  \/   |_| |_|\___| \_\ \/  \__,_|_| |_| |_|_.__/ \___|_|  
"""

print('Bem vindx ao jogo "Descubra qual é o número!"')
print(logo)

#Váriavel com o valor aleatório que o computador escolheu.
numero = random.choice(range(1,100))
encerrar_jogo = False

print("Estou pensando em um número entre 1 e 100")
dificuldade = input("Escolha um grau de dificuldade. Escreva 'fácil', 'intermediário' ou difícil: ")
dificuldade_tentativas = {"fácil": 10, "intermediário": 6, "difícil": 4, "facil": 10, "intermediario": 6, "dificil": 4}

#Loop for para percorrer o dicionário dificuldades_tentativas para definir quantas tentivas o usuário terá direito de acordo com 
for item in dificuldade_tentativas:
    if item == dificuldade:
        tentativas = dificuldade_tentativas[item]
        
#Frases que serão mostradas aleatoriamente a depender do quão próximo o jogador está do número certo       
frases_erro_longe = ['Você errou! O número está beeem longe.', 'Ainda não é esse o número, você está muito distante.',
                     'O número não é esse, você precisa chegar mais perto.', 'Você errou e está muito frio',
                     'Não é esse o número, está muuuito frio']

frases_erro_perto = ['Você errou, mas está próximo', 'Quaaase lá! Você está chegando perto', 'Você está bem perto, hein! Ainda não é o número.',
                     'Pensa mais um pouco, você está muito próximo de acertar', 'O número não é esse, a boa notícia é que você está próximo :)']
                     
frases_erro_muito_perto = ['Eu quase enlouqueci, está muito quente!', 'Está muuuuito quente', 'Você chegou tão perto que eu pensei que fosse acertar',
                          'Não acredito!!! Você está muito perto de adivinhar', 'O palpite é muito bom e você está quase acertando']

while encerrar_jogo == False:
    
    print(f"Você tem {tentativas} tentativas para adivinhar qual é o número")
    palpite = int(input("Faça um palpite: "))
    
    #Variável para criar um indice para puxar da lista uma frase da lista de modo aleatório
    indice = random.choice(range(1,5))
    
    #Calculando a diferença do palpite e do valor correto. A função abs é para deixar o valor sempre positivo.
    diferenca = abs(palpite - numero)
     
    #Condicionais considerando que o usuário erre, e a sua negação que no caso é acertar. Caso o número for diferente, vai mostrar uma frase aleatória e continuar o while e caso as tentativas acabem, o jogo termina.     
    if palpite != numero:
        tentativas -= 1
        if tentativas == 0:
            print(f"Infelizmente as suas tentativas acabaram e você não acertou o número. A resposta certa era {numero}")
            encerrar_jogo = True
        if diferenca < 3:
            print(frases_erro_muito_perto[indice])
        elif diferenca < 10:
            print(frases_erro_perto[indice])        
        else:
            print(frases_erro_longe[indice])
        continue
        
    else:
        print(f'Você conseguiu! O número é {palpite}')
        encerrar_jogo = True    
```

```python
#Geração de modo aletório da palavra que deverá ser adivinhada pelo usuário dentro das possíveis que estão na variável word_list.
chosen_word = random.choice(word_list)

#Criação automática das posições de acordo com a quantidade de letras que existe na palavra.
display = []
for letter in chosen_word:
    display.append('_')

#Quantidade de vidas do usuário e index para que sejam mostrados os stages de acordo com a quantidade de vidas disponíveis.    
lives = 6
index = 0

#Mostrando a logo do jogo, 
print(logo)


#Loop para que, enquanto o usuário não acerte todas as letras o jogo deverá continuar rodando.
while "_" in display:
    
    #Mostrar as posições da palavra, a figura correspondente a quantidade de vida e o campo de preenchimento.
    print(*display, sep = ' ')
    print(stages[index])
    guess = input('Guess a letter: ').lower()
    
    #Caso o usuário não acerte, será retirado uma vida da váriavel lives e adicionado uma posição a variável index para que possa ser printado o boneco correspondete.
    if guess not in chosen_word:
        lives -= 1
        index += 1
        print(f"You guessed {guess}, that's not in the word. You lose a life. Lives: {lives}")
        
    #Caso o usuário tenha acertado a letra, a respectiva letra será substituída no display[]
    for i,letter in enumerate(chosen_word):
        if guess == letter:
            display[i] = letter
            
    #Caso as vidas tenham acabado, o jogador perde o jogo e o game é encerrado.
    if lives == 0:
        print(stages[index])
        print(f'You lose! The answer is {chosen_word}')
        break
#Transformando a lista display em uma string dentro da variável answer          
answer = "".join(display)

#Criação de condicional para que somente mostre a mensagem de que o jogador ganhou quando a variável answer(respostas do usuário) seja igual a chosen_word(escolha randômica do computador).
if answer == chosen_word:
    print(f"You win! The word is {answer}")
```

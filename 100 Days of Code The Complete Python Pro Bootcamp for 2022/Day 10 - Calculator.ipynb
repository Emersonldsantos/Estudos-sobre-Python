Create a calculator.

```python
#Importar a função de apagar os dados da tela da bilioteca replit
from replit import clear
```

```python
# Definindo as funções de soma, subtração, multiplicação, divisão e a própria calculadora.
def plus(x,y):
    result = x + y
    return result

def minus(x,y):
    result = x - y
    return result

def times(x,y):
    result = x * y
    return result

def divided(x,y):
    result = x / y
    return result

def calculator():
    print("""
 _____________________
|  _________________  |
| | Pythonista   0. | |  .----------------.  .----------------.  .----------------.  .----------------. 
| |_________________| | | .--------------. || .--------------. || .--------------. || .--------------. |
|  ___ ___ ___   ___  | | |     ______   | || |      __      | || |   _____      | || |     ______   | |
| | 7 | 8 | 9 | | + | | | |   .' ___  |  | || |     /  \     | || |  |_   _|     | || |   .' ___  |  | |
| |___|___|___| |___| | | |  / .'   \_|  | || |    / /\ \    | || |    | |       | || |  / .'   \_|  | |
| | 4 | 5 | 6 | | - | | | |  | |         | || |   / ____ \   | || |    | |   _   | || |  | |         | |
| |___|___|___| |___| | | |  \ `.___.'\  | || | _/ /    \ \_ | || |   _| |__/ |  | || |  \ `.___.'\  | |
| | 1 | 2 | 3 | | x | | | |   `._____.'  | || ||____|  |____|| || |  |________|  | || |   `._____.'  | |
| |___|___|___| |___| | | |              | || |              | || |              | || |              | |
| | . | 0 | = | | / | | | '--------------' || '--------------' || '--------------' || '--------------' |
| |___|___|___| |___| |  '----------------'  '----------------'  '----------------'  '----------------' 
|_____________________|
"""  )
    
    #Variável que define se a função continuará rodando ou não
    continue_calculating = 'y'
    x = float(input("What's the first number: "))
    
    
    while continue_calculating != "n":    
        print("\n + \n - \n * \n /")
        operation = input("Pick an operation: ")
        y = float(input("What's the next number: "))

        list_operation = ["+","-","*","/"]
        list_functions = [plus(x,y),minus(x,y),times(x,y),divided(x,y)]
        
        #Loof for percorrendo as duas lsistas acima, para selecionar a função correta de acordo com o input do usuário em operation.   
        for i, item in enumerate(list_operation):
            if operation == item:
                result = list_functions[i]
                
        #Mostrar o resultado
        print(str(x) + " " + str(operation) + " " + str(y) + " " + "=" + " " + str(result))

        if input(f"Type 'y' to continue calculating with {result}, or type 'n' to start a new calculation: ") == "y":
            x = result
            continue
        else:
            clear()
            calculator()
```

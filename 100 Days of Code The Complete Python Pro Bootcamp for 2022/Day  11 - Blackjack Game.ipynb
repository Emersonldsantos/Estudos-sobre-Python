# Blackjack Game

#### Our Blackjack House Rules

- The deck is unlimited in size.
- There are no jokers. 
- The Jack/Queen/King all count as 10.
- The the Ace can count as 11 or 1.
- Use the following list as the deck of cards:
- cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
- The cards in the list have equal probability of being drawn.
- Cards are not removed from the deck as they are drawn.
- The computer is the dealer

```python
import random
from replit import clear
```

```python
def deal_card():
    '''Esta fun칞칚o escolhe aleatoriamente uma carta do baralho'''
    import random
    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
    choice = random.choice(cards)
    return choice

def calculate_score(listofcards):
    '''Retorna o score do jogo a partir de uma lista de cartas do jogador'''
    score = sum(listofcards)
    if score == 21 and len(listofcards) == 2:
        score = 0
    if 11 in listofcards and score > 21:
            listofcards.remove(11)
            listofcards.append(1)
            score = sum(listofcards)
    return score


def compare(user_score, computer_score):
    '''Esta fun칞칚o faz a compara칞칚o entre os scores do jogador e do computador e retorna o resultado do jogo'''
    if user_score == computer_score:
        check = "Draw 游뗶"     
    elif computer_score == 0:
        check = 'Lose, opponent has Blackjack 游땸.'        
    elif user_score == 0:
        check = 'Win with a Blackjack 游땙'
    elif user_score > 21:
        check = 'You went over. You lose 游땴'
    elif computer_score > 21:
        check = 'Opponent went over. You win 游때'
    elif computer_score > user_score:
        check = 'You win 游땎'
    else:
        check = 'You lose 游땫'
    return check

def playthegame():

    logo = """
    .------.            _     _            _    _            _    
    |A_  _ |.          | |   | |          | |  (_)          | |   
    |( \/ ).-----.     | |__ | | __ _  ___| | ___  __ _  ___| | __
    | \  /|K /\  |     | '_ \| |/ _` |/ __| |/ / |/ _` |/ __| |/ /
    |  \/ | /  \ |     | |_) | | (_| | (__|   <| | (_| | (__|   < 
    `-----| \  / |     |_.__/|_|\__,_|\___|_|\_\ |\__,_|\___|_|\_\\
          |  \/ K|                            _/ |                
          `------'                           |__/           
    """
    print(logo)

    user_cards = []
    computer_cards = []
    game_ends = False

    #Adicionar as cartas escolhidas aletoriamente as listas de cartas do usu치rio e do computador.
    user_cards.append(deal_card())
    user_cards.append(deal_card())
    computer_cards.append(deal_card())
    computer_cards.append(deal_card())


    while game_ends == False:
        #Calculando e mostrando os scores do usu치rio e do computador
        user_score = calculate_score(user_cards)
        computer_score = calculate_score(computer_cards)    
        print(f"Your cards: {user_cards}, current score: {user_score}")
        print(f"Computer's first card: {computer_cards[0]}")

        #Comparando 
        if user_score == 0 or computer_score == 0 or user_score > 21:
            game_ends = True
        else:
            if input("Type 'y' to get another card, type 'n' to pass: ") == 'y':
                user_cards.append(deal_card())
            else:
                game_ends = True

        #O c칩digo dever치 continuar adicionando cartas a lista do computador enquanto a soma total n칚o for > 17 e 
        #o score diferente de zero.
        while sum(computer_cards) < 17 and computer_score != 0:
            computer_cards.append(deal_card())
            computer_score = calculate_score(computer_cards)

        print(f"   Your final hand: {user_cards}, final score: {user_score}")
        print(f"   Computer's final hand: {computer_cards}, final score: {computer_score}")
        print(compare(user_score, computer_score))

        #Enquanto o usu치rio decidir continuar jogando, a fun칞칚o do jogo ser치 acionada.
    while input("Do you want to play a game of Blackjack? Type 'y' or 'n': ") == "y":
        clear()
        playthegame()
```
